# Dataset Quality and EDA

EDA is used here to justify preprocessing decisions, not just to display generic plots.


```python
from pathlib import Path
import json
import pandas as pd
pd.options.display.max_colwidth = 120
pd.options.display.float_format = lambda x: f"{x:0.4f}"
ROOT = Path.cwd()
FINAL_OUTPUTS = ROOT / "final_outputs"
PROCESSED_DIR = ROOT / "data" / "processed"

summary = json.loads((FINAL_OUTPUTS / "dataset_summary.json").read_text())
display(pd.DataFrame([
    ["Modeling rows", summary["n_modeling_rows"]],
    ["Processed columns", summary["n_processed_columns"]],
    ["Baseline model input features", summary["n_model_input_features_baseline"]],
    ["Overall target rate", summary["target_rate_all"]],
    ["Missing review-text rate", summary["missing_text_rate"]],
], columns=["Item", "Value"]))
```

![Class imbalance](../submission/figures/eda_class_imbalance.png)

The target is rare in every split, so accuracy would be misleading.

![Missingness](../submission/figures/eda_missingness.png)

Review text is missing for a large share of customers, which explains why text-only models are limited.

![Skew before and after transforms](../submission/figures/eda_skew_before_after.png)

Strong right skew in price, freight, and delivery variables justifies log transforms.

![Temporal cohort](../submission/figures/eda_temporal_cohort.png)

The score-time cohort changes over time, so chronological splitting is safer than a random split.


Leakage is controlled by using only the first delivered order for each customer and defining `score_time` as the later of delivery and review creation time.
