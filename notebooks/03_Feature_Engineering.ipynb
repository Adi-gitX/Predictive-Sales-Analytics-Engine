# Feature Engineering

Feature engineering turns the raw relational tables into a customer-level retention problem.


```python
from pathlib import Path
import json
import pandas as pd
pd.options.display.max_colwidth = 120
pd.options.display.float_format = lambda x: f"{x:0.4f}"
ROOT = Path.cwd()
FINAL_OUTPUTS = ROOT / "final_outputs"
PROCESSED_DIR = ROOT / "data" / "processed"

feature_df = pd.DataFrame([
    ["Review features", "review_score, text_present, text_char_len, text_word_count", "Customer satisfaction and engagement"],
    ["Monetary features", "total_price, total_freight, payment_value_total, freight_ratio, payment_gap", "Economic burden and checkout behavior"],
    ["Order complexity", "item_count, seller_count", "Fulfillment complexity"],
    ["Delivery features", "approval lag, delivery days, clipped delivery delay, late flag", "Operational service quality"],
    ["Categorical context", "customer_state, seller_state_mode, payment_type_mode, product_category_main", "Regional and product context"],
], columns=["Feature family", "Examples", "Why it matters"])
display(feature_df)
```
