# Model Application and Evaluation

This phase-1 notebook compares only baseline models so the final choice remains simple and defensible.


```python
from pathlib import Path
import json
import pandas as pd
pd.options.display.max_colwidth = 120
pd.options.display.float_format = lambda x: f"{x:0.4f}"
ROOT = Path.cwd()
FINAL_OUTPUTS = ROOT / "final_outputs"
PROCESSED_DIR = ROOT / "data" / "processed"

metrics = pd.read_csv(FINAL_OUTPUTS / "metrics_baselines.csv")
best = json.loads((FINAL_OUTPUTS / "best_model_summary.json").read_text())
display(metrics[metrics["split"] == "test"].sort_values("pr_auc", ascending=False)[["model", "pr_auc", "roc_auc", "precision_at_k", "lift_at_k", "threshold"]])
display(pd.DataFrame([
    ["Best validation baseline", best["overall_validation_best_model"]],
    ["Best test baseline", best["overall_test_best_model"]],
    ["Recommended final baseline", best["recommended_model"]],
], columns=["Decision point", "Model"]))
```
