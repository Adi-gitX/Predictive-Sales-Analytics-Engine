# Predictive Sales Analytics Engine

This notebook presents the final phase-1 baseline submission: leakage-safe dataset construction, clean EDA, baseline model comparison, and one simple final recommendation.


```python
from pathlib import Path
import json
import pandas as pd
pd.options.display.max_colwidth = 120
pd.options.display.float_format = lambda x: f"{x:0.4f}"
ROOT = Path.cwd()
FINAL_OUTPUTS = ROOT / "final_outputs"
PROCESSED_DIR = ROOT / "data" / "processed"

raw_files = sorted([p.name for p in (ROOT / "data" / "raw").glob("*.csv")])
summary = json.loads((FINAL_OUTPUTS / "dataset_summary.json").read_text())
best = json.loads((FINAL_OUTPUTS / "best_model_summary.json").read_text())
metrics = pd.read_csv(FINAL_OUTPUTS / "metrics_baselines.csv")
display(pd.DataFrame({"raw_tables": raw_files}))
display(pd.DataFrame([
    ["Modeling rows", summary["n_modeling_rows"]],
    ["Baseline model input features", summary["n_model_input_features_baseline"]],
    ["Overall target rate", summary["target_rate_all"]],
    ["Recommended final model", best["recommended_model"]],
    ["Test PR-AUC", best["recommended_model_test_metrics"]["pr_auc"]],
    ["Test Lift@10%", best["recommended_model_test_metrics"]["lift_at_k"]],
], columns=["Item", "Value"]))
display(metrics[metrics["split"] == "test"].sort_values("pr_auc", ascending=False)[["model", "pr_auc", "roc_auc", "precision_at_k", "lift_at_k"]])
```
